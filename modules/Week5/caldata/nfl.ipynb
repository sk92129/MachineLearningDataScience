```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dictionary with the 49ers' 2023 season statistics
# Replace these placeholder values with actual data
niners_stats = {
    'Week': list(range(1, 18)),  # Assuming a 17-game season
    'Opponent': ['Team' + str(i) for i in range(1, 18)],
    'Result': ['W', 'W', 'L', 'W', 'W', 'W', 'L', 'W', 'W', 'W', 'W', 'L', 'W', 'W', 'W', 'W', 'W'],
    'Passing_Yards': [250, 300, 220, 280, 310, 270, 200, 330, 290, 280, 320, 190, 300, 310, 290, 270, 320],
    'Rushing_Yards': [120, 150, 100, 140, 130, 160, 90, 170, 140, 150, 180, 80, 160, 170, 150, 140, 190],
    'Turnovers': [0, 1, 3, 1, 0, 1, 2, 0, 1, 1, 0, 3, 1, 0, 1, 1, 0]
}

# Create a DataFrame
df = pd.DataFrame(niners_stats)

# Calculate season totals
wins = df['Result'].value_counts()['W']
losses = df['Result'].value_counts()['L']
total_passing_yards = df['Passing_Yards'].sum()
total_rushing_yards = df['Rushing_Yards'].sum()
total_turnovers = df['Turnovers'].sum()

# Print season summary
print(f"San Francisco 49ers 2023 Season Summary")
print(f"Wins: {wins}")
print(f"Losses: {losses}")
print(f"Total Passing Yards: {total_passing_yards}")
print(f"Total Rushing Yards: {total_rushing_yards}")
print(f"Total Turnovers: {total_turnovers}")

# Create a bar plot for wins and losses
plt.figure(figsize=(10, 6))
plt.bar(['Wins', 'Losses'], [wins, losses])
plt.title('49ers 2023 Season: Wins vs Losses')
plt.ylabel('Number of Games')
plt.show()

# Create a line plot for passing and rushing yards per game
plt.figure(figsize=(12, 6))
plt.plot(df['Week'], df['Passing_Yards'], label='Passing Yards')
plt.plot(df['Week'], df['Rushing_Yards'], label='Rushing Yards')
plt.title('49ers 2023 Season: Passing and Rushing Yards per Game')
plt.xlabel('Week')
plt.ylabel('Yards')
plt.legend()
plt.show()

# Create a bar plot for turnovers per game
plt.figure(figsize=(12, 6))
plt.bar(df['Week'], df['Turnovers'])
plt.title('49ers 2023 Season: Turnovers per Game')
plt.xlabel('Week')
plt.ylabel('Number of Turnovers')
plt.show()
```
