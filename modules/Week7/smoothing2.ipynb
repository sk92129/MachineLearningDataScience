```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Generate sample data
n_samples = 1000
age_data = np.concatenate([
    np.random.normal(5, 2, int(n_samples * 0.05)),   # Young children
    np.random.normal(15, 3, int(n_samples * 0.1)),   # Teenagers
    np.random.normal(30, 8, int(n_samples * 0.4)),   # Young adults
    np.random.normal(50, 10, int(n_samples * 0.35)), # Middle-aged adults
    np.random.normal(75, 5, int(n_samples * 0.1))    # Elderly
])

# Clip ages to be between 0 and 80
age_data = np.clip(age_data, 0, 80)

# Create DataFrame
df = pd.DataFrame({'age': age_data})

# Create the plot
plt.figure(figsize=(10, 6))

# Create a histogram with KDE
# the kde parameter controls whether to plot a KDE
sns.histplot(data=df, x='age', kde=True, bins=30, color='skyblue', edgecolor='white')

# Customize the plot
plt.title('Age Distribution with KDE', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Remove top and right spines
sns.despine()

# Show the plot
plt.tight_layout()
plt.show()

# Display first few rows of the DataFrame
print(df.head())

# Display summary statistics
print(df.describe())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Generate sample data
n_samples = 1000
age_data = np.concatenate([
    np.random.normal(5, 2, int(n_samples * 0.05)),   # Young children
    np.random.normal(15, 3, int(n_samples * 0.1)),   # Teenagers
    np.random.normal(30, 8, int(n_samples * 0.4)),   # Young adults
    np.random.normal(50, 10, int(n_samples * 0.35)), # Middle-aged adults
    np.random.normal(75, 5, int(n_samples * 0.1))    # Elderly
])

# Clip ages to be between 0 and 80
age_data = np.clip(age_data, 0, 80)

# Create DataFrame
df = pd.DataFrame({'age': age_data})

# Create the plot
plt.figure(figsize=(10, 6))

# Create a histogram with KDE
# the kde parameter controls whether to plot a KDE

sns.kdeplot(data=df, x='age', color='darkblue')

# Customize the plot
plt.title('Age Distribution with KDE', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Remove top and right spines
sns.despine()

# Show the plot
plt.tight_layout()
plt.show()

# Display first few rows of the DataFrame
print(df.head())

# Display summary statistics
print(df.describe())
```

More hyper parameters to adjust and control the KDE curve

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming we still have our 'df' DataFrame with the 'age' column

plt.figure(figsize=(12, 6))

# Histogram with default KDE
sns.histplot(data=df, x='age', kde=True, color='skyblue', label='Default KDE', alpha=0.5)

# Additional KDE curves with different parameters
sns.kdeplot(data=df, x='age', bw_adjust=0.5, color='red', label='Less Smooth (bw_adjust=0.5)')
sns.kdeplot(data=df, x='age', bw_adjust=2, color='green', label='More Smooth (bw_adjust=2)')
sns.kdeplot(data=df, x='age', kernel='cos', color='purple', label='Cosine Kernel')
sns.kdeplot(data=df, x='age', cumulative=True, color='orange', label='Cumulative')

plt.title('Age Distribution with Various KDE Parameters', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```
