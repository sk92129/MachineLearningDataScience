```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

# Sample data: list of GPS points (latitude, longitude)
gps_points = [(37.8715, -122.27), (37.869, -122.25), (37.874, -122.28)]

# Create a GeoDataFrame from the GPS points
gdf = gpd.GeoDataFrame(geometry=gpd.points_from_xy([point[1] for point in gps_points], [point[0] for point in gps_points]))

# Set the coordinate reference system to match contextily's (Web Mercator)
gdf = gdf.set_crs(epsg=4326)
gdf = gdf.to_crs(epsg=3857)

# Plot the GPS points
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, marker='o', color='red', markersize=50)

# Add a basemap
ctx.add_basemap(ax, source=ctx.providers.Esri.WorldImagery)

# Set plot title and labels
plt.title('GPS Points in Berkeley, California')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show plot
plt.show()
```
