```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data for virginica
virginica_sepal_width = np.random.normal(3.0, 0.3, 500)
virginica_sepal_length = np.random.normal(6.7, 0.5, 500)

# Generate synthetic data for setosa
setosa_sepal_width = np.random.normal(3.5, 0.3, 500)
setosa_sepal_length = np.random.normal(5.0, 0.3, 500)

# Create a DataFrame
df = pd.DataFrame({
    'sepal_width': np.concatenate([virginica_sepal_width, setosa_sepal_width]),
    'sepal_length': np.concatenate([virginica_sepal_length, setosa_sepal_length]),
    'species': ['virginica'] * 500 + ['setosa'] * 500
})

# Create the plot
plt.figure(figsize=(10, 8))
sns.kdeplot(data=df, x='sepal_width', y='sepal_length', hue='species', 
             fill=True,  cmap='coolwarm', alpha=0.7, cbar=False)

# Customize the plot
plt.title('Flower Data', fontsize=16)
plt.xlabel('sepal_width', fontsize=12)
plt.ylabel('sepal_length', fontsize=12)

# Adjust the axis limits to match the image
plt.xlim(2.0, 4.5)
plt.ylim(4.0, 8.5)

# Show the plot
plt.tight_layout()
plt.show()
```
