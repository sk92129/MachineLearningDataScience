```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a sample dataframe
np.random.seed(0)
size = 100
X = np.random.normal(loc=0, scale=1, size=size)
Y = np.random.normal(loc=0, scale=1, size=size)

df = pd.DataFrame({'X': X, 'Y': Y})

# Step 2: Apply transformations to demonstrate the Bulge Diagram theory
df['Y_sqrt'] = np.sqrt(df['Y'] - df['Y'].min() + 1)  # Square root transformation
df['Y_log'] = np.log(df['Y'] - df['Y'].min() + 1)    # Log transformation
df['Y_square'] = df['Y'] ** 2                        # Square transformation

# Step 3: Plot the data
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Original data
axs[0, 0].scatter(df['X'], df['Y'])
axs[0, 0].set_title('Original Data')

# Square root transformation
axs[0, 1].scatter(df['X'], df['Y_sqrt'])
axs[0, 1].set_title('Square Root Transformation')

# Log transformation
axs[1, 0].scatter(df['X'], df['Y_log'])
axs[1, 0].set_title('Log Transformation')

# Square transformation
axs[1, 1].scatter(df['X'], df['Y_square'])
axs[1, 1].set_title('Square Transformation')

plt.tight_layout()
plt.show()
```
