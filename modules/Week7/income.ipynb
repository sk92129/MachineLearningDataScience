```python
import pandas as pd
import numpy as np

# Create the data
data = {
    'Education': ['<HS', 'HS', '<BA', 'BA', '>BA'],
    'Men': [520, 750, 880, 1250, 1390],
    'Women': [410, 580, 660, 970, 1050]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Calculate the M/F Income Ratio
df['M/F Ratio'] = df['Men'] / df['Women']

# Display the DataFrame
print(df)

# Display summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming we already have the DataFrame 'df' from the previous code

# Set the style for the plot



# Create the plot
plt.figure(figsize=(10, 6))

# Plot lines for Men and Women
plt.plot('Education', 'Men', data=df, marker='o', color='blue', linewidth=2, label='Men')
plt.plot('Education', 'Women', data=df, marker='o', color='red', linewidth=2, label='Women')

# Customize the plot
plt.title('2014 Median Weekly Earnings\nFull-Time Workers over 25 years old', fontsize=14)
plt.xlabel('Education', fontsize=12)
plt.ylabel('Income', fontsize=12)
plt.legend(title='gender', loc='upper left', fontsize=10)

# Set y-axis limits to match the image
plt.ylim(400, 1500)

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming we already have the DataFrame 'df' from the previous code

# Set the seaborn style
sns.set_style("whitegrid")

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the M/F Ratio line
sns.lineplot(x='Education', y='M/F Ratio', data=df, marker='o', color='blue', linewidth=2)

# Customize the plot
plt.title('M/F Income Ratio as a function of education level', fontsize=14)
plt.xlabel('Education', fontsize=12)
plt.ylabel('Ratio', fontsize=12)

# Set y-axis limits to match the image
plt.ylim(1.26, 1.35)

# Adjust y-axis ticks to match the image
plt.yticks([1.27, 1.28, 1.29, 1.30, 1.31, 1.32])

# Show the plot
plt.tight_layout()
plt.show()
```
