```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Create the first dataframe: Skewed distribution
df1 = pd.DataFrame({
    'Group': 'A',
    'Birth Weight': np.concatenate([
        np.random.normal(120, 10, 500),  # Main cluster
        np.random.normal(150, 5, 100),   # Secondary cluster
        np.random.normal(90, 8, 50)      # Small cluster of lower weights
    ])
})

# Create the second dataframe: Bimodal distribution
df2 = pd.DataFrame({
    'Group': 'B',
    'Birth Weight': np.concatenate([
        np.random.normal(110, 8, 300),   # First mode
        np.random.normal(130, 8, 300),   # Second mode
        np.random.uniform(80, 160, 100)  # Uniform spread for some variability
    ])
})

# Combine the dataframes
df_combined = pd.concat([df1, df2], ignore_index=True)

# Display the first few rows of the combined dataframe
print(df_combined.head(10))

# Display summary statistics
print(df_combined.groupby('Group')['Birth Weight'].describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.violinplot(x='Group', y='Birth Weight', 
                hue='Group',
               data=df_combined, 
               palette={'A': sns.color_palette()[0], 'B': 'red'})
plt.title('Birth Weight Distribution by Group')
plt.show()
```

you can see above that the left side blue violin is unimodal, where as the right side red violin is bimodal.
