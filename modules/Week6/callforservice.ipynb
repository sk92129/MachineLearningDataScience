```python
import folium

# Create a map centered on Berkeley, California
berkeley_map = folium.Map(location=[37.8715, -122.2730], zoom_start=13)

# Define some points of interest around Berkeley
points_of_interest = [
    {"name": "UC Berkeley", "location": [37.8719, -122.2585]},
    {"name": "Berkeley Marina", "location": [37.8645, -122.3130]},
    {"name": "Berkeley Rose Garden", "location": [37.8832, -122.2638]},
    {"name": "Berkeley Bowl", "location": [37.8478, -122.2732]},
    {"name": "Tilden Regional Park", "location": [37.8981, -122.2474]}
]

# Add markers for each point of interest
for poi in points_of_interest:
    folium.Marker(
        location=poi["location"],
        popup=poi["name"],
        tooltip=poi["name"]
    ).add_to(berkeley_map)

# Display the map
berkeley_map
```
