```python
import pandas as pd

data = {
    'x': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y': [ 8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68],
}

df1 = pd.DataFrame(data)

print(df1)
```

```python
data2 = {
    'x': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y': [ 9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74],
}

df2 = pd.DataFrame(data2)

print(df2)
```

```python
data3 = {
    'x': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y': [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73],
}

df3 = pd.DataFrame(data3)

print(df3)
```

```python
data4 = {
    'x': [8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 19.0, 8.0, 8.0, 8.0],
    'y': [6.56, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89],
}

df4 = pd.DataFrame(data4)

print(df4)
```

```python

for i, df in enumerate([df1, df2, df3, df4], 1):
    x_avg = df['x'].mean()
    y_avg = df['y'].mean()
    x_std = df['x'].std()
    y_std = df['y'].std()
    correlation = df['x'].corr(df['y'])


  
    print(f"\nDataFrame {i} Statistics:")
    print(f"Average x: {x_avg:.2f}")
    print(f"Average y: {y_avg:.2f}")
    print(f"Standard deviation x: {x_std:.2f}")
    print(f"Standard deviation y: {y_std:.2f}")
    print(f"Correlation between x and y: {correlation:.2f}")
```

```python
import matplotlib.pyplot as plt
# Create Anscombe's quartet plot
fig, axs = plt.subplots(2, 2, figsize=(12, 12))
fig.suptitle("Anscombe's Quartet")

dataframes = [df1, df2, df3, df4]
titles = ['Dataset 1', 'Dataset 2', 'Dataset 3', 'Dataset 4']

for i, (df, title, ax) in enumerate(zip(dataframes, titles, axs.flatten())):
    ax.scatter(df['x'], df['y'])
    ax.set_title(title)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_xlim(0, 20)
    ax.set_ylim(0, 15)
    
    # Add text with statistics
    stats_text = f"Mean x: {df['x'].mean():.2f}\n" \
                 f"Mean y: {df['y'].mean():.2f}\n" \
                 f"Std x: {df['x'].std():.2f}\n" \
                 f"Std y: {df['y'].std():.2f}\n" \
                 f"Corr: {df['x'].corr(df['y']):.2f}"
    ax.text(0.05, 0.95, stats_text, transform=ax.transAxes, 
            verticalalignment='top', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

data = {
    'Flavor': ['Chocolate', 'Strawberry', 'Vanilla'],
    'Number_of_Cartons': [15, 5, 8]
}

df = pd.DataFrame(data)

print(df)
```

```python

# Create the bar plot
plt.figure(figsize=(10, 6))
bars = plt.barh(df['Flavor'], df['Number_of_Cartons'], color='#4C72B0')

# Customize the plot
plt.title('Number of Cartons by Flavor', fontsize=14)
plt.xlabel('Number of Cartons', fontsize=12)
plt.ylabel('Flavor', fontsize=12)

# Add value labels at the end of each bar
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2, f'{width}', 
             ha='left', va='center', fontweight='bold')

# Adjust the layout and display the plot
plt.xlim(0, 16)  # Set x-axis limit to match the image
plt.gca().invert_yaxis()  # Invert y-axis to match the image order
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
