```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

# Generate skewed right data
np.random.seed(42)
data = skewnorm.rvs(a=5, loc=5, scale=2, size=1000)

# Create a DataFrame
df = pd.DataFrame({'values': data})

# Create the plot
plt.figure(figsize=(10, 6))
plt.hist(df['values'], bins=30, edgecolor='black')
plt.title('Skewed Right Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Add mean and median lines
plt.axvline(df['values'].mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(df['values'].median(), color='green', linestyle='dashed', linewidth=2, label='Median')

plt.legend()
plt.show()
```
